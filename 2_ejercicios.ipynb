# Analizando una Imagen

```python
import matplotlib.pyplot as plt 
img = plt.imread('beagle.jpg')
img.shape

img[10,10]
```

```python
print(f"Dimensiones: {img.shape}")
print(f"Alto: {img.shape[0]} píxeles")
print(f"Ancho: {img.shape[1]} píxeles") 
print(f"Canales RGB: {img.shape[2]}")
print(f"Tipo de dato: {img.dtype}")
```

```python
print(f"Valor del píxel en [10, 10]: {img[10, 10]}")
print(f"Valor del píxel en [500, 500]: {img[500, 500]}")
print("Rango de valores: 0 a 255")
```
